```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("SU_building.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Copy image
img_text = img.copy()

# Add text
cv2.putText(img_text, "SNU", (50,50),
            cv2.FONT_HERSHEY_SIMPLEX,
            1, (255,0,0), 2)

# Show both images
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(img_text)
plt.title("Sreenidhi Univesity")

plt.show()
```
