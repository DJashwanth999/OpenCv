```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(r"C:\Users\USER\Downloads\IMG20260625100140.jpg.jpeg")

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Edge detection
edges = cv2.Canny(gray, 40, 55, apertureSize=3)

# Invert edges
edges_inv = cv2.bitwise_not(edges)

# Convert edge image to 3 channels for blending
edges_color = cv2.cvtColor(edges_inv, cv2.COLOR_GRAY2BGR)

# Blend images
blended = cv2.addWeighted(edges_color, 0.3, image, 0.7, 0)

# Convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
blended_rgb = cv2.cvtColor(blended, cv2.COLOR_BGR2RGB)

# Display side by side
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title("Original Palm")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(edges_inv, cmap="gray")
plt.title("Edge Detected Palm")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(blended_rgb)
plt.title("Blended Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
